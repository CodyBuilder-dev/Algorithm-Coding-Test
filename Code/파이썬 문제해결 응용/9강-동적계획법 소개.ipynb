9-1. 피보나치 수열

```python
'''
목표 : 피보나치 수열을 재귀적으로 구하는 함수를 작성하라
예시 :
제약조건 : 1. 스스로 작성할 것
'''
```

```python
"내 코드1 - 피보나치 재귀 "
#아이디어 : f(0) = 0, f(1) = 1, f(2) = 1, f(3) = 2...
#즉, f(0),f(1)은 정의해 줘야 함

#시간측정용
import timeit

#함수 정의
def fibonacci(n):
    if n == 0 : return 1
    elif n ==1 : return 1
    else :
        return fibonacci(n-1)+fibonacci(n-2)

#시간측정용
start = timeit.default_timer()
fibonacci(1000)
stop = timeit.default_timer()
print(stop - start)
```

9-3 메모이제이션

```python
'''
목표 : 피보나치 수열을 재귀적으로 구하되, 메모이제이션 쓸 수 있도록 하기
예시 :
제약조건 : 1. 스스로 작성할 것
'''
```

```python
"내 코드2 - 피보나치 메모이제이션"
#아이디어 : 호출을 하기 전에, 메모리 되있나 생각해보기

#시간측정용
import time
import timeit

memo_list = dict()
#함수 본체
def fibonacci_memo(n):
    if n == 0 :
        if 0 in memo_list : return 1
        else : memo_list[0] = 1; return memo_list[0]
        
    elif n ==1 :
        if 1 in memo_list : return 1
        else : memo_list[1] = 1; return memo_list[1]
    else :
        if n-1 in memo_list : pass
        else : memo_list[n-1] = fibonacci_memo(n-1)
        if n-2 in memo_list : pass
        else : memo_list[n-2] = fibonacci_memo(n-2)
        return memo_list[n-1] + memo_list[n-2]

#시간측정용
start = timeit.default_timer()
fibonacci_memo(1000)
stop = timeit.default_timer()
print(stop - start)
"문제점 : 그냥 코드보다 느려지진 않는데, 더럽게 복잡함"
```

```python
"예제코드1 - 피보나치 메모이제이션"
import timeit

memo = [0]*10000
memo[0] = 0
memo[1] = 1

def fibo1(n):

    if n>=2 and memo[n] == 0:
        memo[n] = fibo1(n-1)+fibo1(n-2)
    return memo[n]

start = timeit.default_timer()
fibo1(1000)
stop = timeit.default_timer()
print(stop - start)

"문제점 : 처음에 memo를 초기화해줘야 함"
```
