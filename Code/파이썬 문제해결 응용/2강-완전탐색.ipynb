```python
'''
목표 : 입력을 받아 리스트에서 원하는 값을 찾는 함수
입력 : a(리스트) , n(리스트 길이?), key

출력 : 성공시 1 실패시 -1
제한조건 : 1. 직접 작성할 것
'''
```

```python
"내 코드"

#전략 : 리스트 길이가 유한하므로, for문 돌려서 찾자

def sequentialSearch(a,n,key):
    for i in range(n):
        if a[i] == key : return 1
    return -1
```

```python
"예제 코드"
def sequentialSearch(a,n,key):
    i =0
    while i<n and a[i] != key :
        i = i+1
    if i<n : return 1
    else : return -1
```
